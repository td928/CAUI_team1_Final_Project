## Data Ingestion and Processing

```python
# Import
import numpy as np
import pandas as pd
from datetime import datetime

from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.mixture import GaussianMixture
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore')

import matplotlib as plt
import matplotlib.style as style
from pandas.tools.plotting import scatter_matrix
style.use('fivethirtyeight')
%pylab inline

from IPython.display import HTML, display as d

pd.set_option('display.max_columns', 500)
```

```python
# Ingest data from project github
# Please see Data Processing script for detailed steps (TO BE UPLOADED)
url = "https://raw.githubusercontent.com/td928/CAUI_team1_Final_Project/master/merged-w-latlon.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.dtypes
```

```python
# Create a Age feature based on Built Year
df["age"] = datetime.datetime.now().year - df['YearBuilt']

# Create Average EUI of 2013, 2014, and 2015
df["avgEUI"] = df[['EUI_2013', 'EUI_2014', 'EUI_2015']].mean(axis=1)

df.head(10)
```

```python
# Create a unique list of BBL

uni_BBL = df[["BBL", "Longitude", "Latitude", "age", "BldgArea", "UnitsTotal"]].drop_duplicates()
uni_BBL.shape
```

```python
print("--- Number of Missing Value in Each Column ---")
d(df.isnull().sum())
len_before = len(df)

# Drop rows with any missing value
model_df = df.dropna()
len(model_df) - len_before, 'dropped'
```

```python
# Select only Multi-family buildings for Modelling 

model_df = model_df[(model_df.TYPE_2016 == 'Multifamily Housing')]
print("Number of Records in the Modelling Dataset: {}".format(model_df.shape[0]))
```

## Modelling
- Calculate EER, WER, NER using K-Means
- Calculate EER, WER, NER using Linear Regression

EER Calculation:
- In K-Means, EER = actual EUI 2016 / K-mean Group EUI 2016.
- In Linear Regression, EER = actual EUI 2016 / Predicted EUI 2016

```python
model_df.dtypes
```

```python
# Select K-Means Features

kmean_features = ['age', 'Occupancy', 'Latitude', 'Longitude', 'BldgArea', 'UnitsTotal', 'avgEUI']
seed = 123
```

```python
# Find the optimal K 

def findK(df, kmean_features, k_range):
    
    silhoutte_score = []
    
    for n in range(2, k_range):
        kmeans = KMeans(n_clusters = n, random_state = 123, n_init = 100)
        km_label = kmeans.fit_predict(df)
        silhouette_avg = silhouette_score(df, km_label)
        silhoutte_score.append(silhouette_avg)
        print("Cluster = %i | Silhouette Score: %.3f" % (n, silhouette_avg))
    
    n_range = range(2, k_range)
    fig = plt.figure(figsize = (8, 8));
    ax = fig.add_subplot(111);
    ax.plot(n_range, silhoutte_score, 'b*-');
    plt.grid(True);
    plt.xlabel('Number of Clusters');
    plt.ylabel('Silhouette Score');
    plt.title('Silhouette Score of A Range of Clusters (Higer the Better)');
    
    return silhoutte_score.index(max(silhoutte_score))+2

k = findK(model_df[kmean_features], kmean_features, 8)

print(" --- the optimal number of cluster: {} ---".format(k))
```

```python
# Perform K-Means using Optimal Number of Cluster
seed = 123
kmeans = KMeans(n_clusters=k, random_state = seed, n_init = 100).fit(model_df[kmean_features])
model_df['km_group'] = kmeans.predict(model_df[kmean_features])
```

```python
model_df.groupby('km_group').size()
```

```python
# Calculate k-mean based EUI statS
km_stats = pd.pivot_table(model_df, index= 'km_group', values= ["EUI_2016", "water_intensity", "naturalgas_intensity"],
                               aggfunc= [np.mean, np.median, min, max, np.std, np.size])

# Rename Columns by Joining with "_"
km_stats.columns = ["Group_"+"_".join((i,j)) for i,j in km_stats.columns]
km_stats = km_stats.reset_index()

km_stats
```

```python
# Merge with Original Modelling Dataset
km_df = model_df.merge(km_stats, how="left", on = "km_group")
```

```python
km_df.shape
```

```python
model_df.shape
```

```python
km_df.sample(10)
```

```python
km_df['EER_km'] = km_df['EUI_2016'] / km_df['Group_mean_EUI_2016']
km_df['WER_km'] = km_df['water_intensity'] / km_df['Group_mean_water_intensity']
km_df['NER_km'] = km_df['naturalgas_intensity'] / km_df['Group_mean_naturalgas_intensity']
km_df[['EER_km', 'WER_km', 'NER_km']].describe()
```

```python
# Remove Unnecessary Columns

keep_col = ["BBL", "EER_km", "WER_km", "NER_km", "EUI_2016", 
               "Group_mean_EUI_2016", "Group_mean_water_intensity", "Group_mean_naturalgas_intensity"]

km_df = km_df[keep_col]
```

```python
# Remove Duplicated BBL by doing an aggregation and pick the max EUI 2016 and EERs
uni_km_df = km_df.groupby("BBL")[keep_col].max()
uni_km_df.shape
```

```python
uni_km_df.dtypes
```

# LINEAR REGRESSION

input dataset: use model_df

output dataset: name it lr_df with on the following columns:
- BBL
- Pred_EER
- EER_lr = (EUI_2016/ Pred_EER) <- not needed based on our discussion, but let's just do it in case we need it

```python
lr_years = [2013, 2014, 2015, 2016]
lr_features = ['age', 'Latitude', 'Longitude', 'BldgArea', 'UnitsTotal']

lr_df = model_df[['BBL']].copy()

# Train model for electricity
# Use EUI for all years
X = pd.concat([model_df[lr_features] for _ in lr_years]) # Duplicate len(years) times
y = pd.concat([model_df['EUI_{}'.format(y)] for y in lr_years]) # Get EUI for each year
lrmodel = LinearRegression().fit(X, y)

# Store predictions
lr_df['Pred_EUI_lr'] = lrmodel.predict(model_df[lr_features])
lr_df['EER_lr'] = model_df['EUI_2016'] / lr_df['Pred_EUI_lr']

# Train model for water
lrmodel = LinearRegression().fit(model_df[lr_features], model_df['water_intensity'])

# Store predictions
lr_df['Pred_WUI_lr'] = lrmodel.predict(model_df[lr_features])
lr_df['WER_lr'] = model_df['water_intensity'] / lr_df['Pred_WUI_lr']

# Train model for natural gas
lrmodel = LinearRegression().fit(model_df[lr_features], model_df['naturalgas_intensity'])

# Store predictions
lr_df['Pred_NUI_lr'] = lrmodel.predict(model_df[lr_features])
lr_df['NER_lr'] = model_df['naturalgas_intensity'] / lr_df['Pred_NUI_lr']

# Retrieve EUI 2016
lr_df["EUI_2016"] = model_df["EUI_2016"]
lr_df.head()
```

```python
# Remove Duplicated BBL by doing an aggregation and pick the max EUI 2016 and EERs
col = lr_df.columns
uni_lr_df = lr_df.groupby("BBL")[col].max()
uni_lr_df.shape
```

```python
# Histogram of Electicity, Hot Water, and Natural Gas EER

fig = plt.figure(figsize = (15,10))
fig.subplots_adjust(hspace=.5)

fig.add_subplot(311)
hist(final_df[(final_df.avgEER > -2) & (final_df.avgEER < 2)].avgEER, bins = 100);
plt.xlabel("Avg. EER (Utility)", size = 12);
plt.ylabel("Count", size = 12);
plt.title("Histogram of Avg. EER (Utility)");
plt.xlim([-0.5, 1.5]);

fig.add_subplot(312)
hist(final_df[(final_df.avgWER > -2) & (final_df.avgWER < 2)].avgWER, bins = 100);
plt.xlabel("Avg. WER (Water)", size = 12);
plt.ylabel("Count", size = 12);
plt.title("Histogram of Avg. WER (Water)");
plt.xlim([-0.5, 1.5]);

fig.add_subplot(313)
hist(final_df[(final_df.avgNER > -2) & (final_df.avgNER < 2)].avgNER, bins = 100);
plt.xlabel("Avg. NER (Gas)", size = 12);
plt.ylabel("Count", size = 12);
plt.title("Histogram of Avg. NER (Gas)");
plt.xlim([-0.5, 1.5]);
```

```python
# Scatter plot between Electicity, Hot Water, and Natural Gas EER

fig = plt.figure(figsize = (15,10))
fig.subplots_adjust(hspace=.3)

fig.add_subplot(131)
plot(log_final_df.avgEER, log_final_df.avgWER, "o", alpha = 0.3);
plt.xlabel("Avg. EER (Utility) in log", size = 12);
plt.ylabel("Avg. WER (Water) in log", size = 12);
plt.title("Utility vs. Water Performance (log scale)")
plt.ylim([-3, 3]);

fig.add_subplot(132)
plot(log_final_df.avgEER, log_final_df.avgNER, "o", alpha = 0.3);
plt.xlabel("Avg. EER (Utility) in log", size = 12);
plt.ylabel("Avg. NER (Gas) in log", size = 12);
plt.title("Utility vs. Gas Performance (log scale)")
plt.ylim([-3, 3]);

fig.add_subplot(133)
plot(log_final_df.avgWER, log_final_df.avgNER, "o", alpha = 0.3);
plt.xlabel("Avg. WER (Water) in log", size = 12);
plt.ylabel("Avg. NER (Gas) in log", size = 12);
plt.title("Water vs. Gas Performance (log scale)");
plt.ylim([-3, 3]);
```
